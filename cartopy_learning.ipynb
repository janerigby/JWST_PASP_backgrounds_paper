mapping projection tutorial from https://scitools.org.uk/cartopy/docs/latest/tutorials/understanding_transform.html

```python
import numpy as np
import cartopy.crs as ccrs
data_crs = ccrs.PlateCarree()
import matplotlib.pyplot as plt
```

```python
lon = np.linspace(-80, 80, 50)
lat = np.linspace(30, 70, 100)
lon2d, lat2d = np.meshgrid(lon, lat)

data = np.cos(np.deg2rad(lat2d) * 4) + np.sin(np.deg2rad(lon2d) * 4)
```

```python
# The projection keyword determines how the plot will look
projection = ccrs.RotatedPole(pole_longitude=-177.5, pole_latitude=37.5)
plt.figure(figsize=(6, 3))
ax = plt.axes(projection=projection)
ax.set_global()
ax.coastlines()

ax.contourf(lon, lat, data, transform=data_crs)  # didn't use transform, but looks ok...
plt.show()
```

```python
projection = ccrs.Mollweide()
plt.figure(figsize=(6, 3))
ax = plt.axes(projection=projection)
ax.set_global()
ax.coastlines()

ax.contourf(lon, lat, data, transform=data_crs) 
plt.show()
```
